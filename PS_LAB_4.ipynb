```python
import pandas as pd
file_path = '/content/car dataset.csv'
car_data= pd.read_csv(file_path)

print(car_data.head())
```

```python
import pandas as pd
import matplotlib.pyplot as plt

plt.hist(car_data['Price'], bins=10, color='red', edgecolor='green')
plt.title('Price Distribution of Cars')
plt.xlabel('Price (USD)')
plt.ylabel('Number of Cars')
plt.show()  # Display the plot
price_range = pd.cut(car_data['Price'], bins=10)


most_common_price_range = price_range.value_counts().idxmax()
print(f"The most common price range is: {most_common_price_range}")
```

```python
import matplotlib.pyplot as plt

# Count the number of cars per brand
brand_counts = car_data['Brand'].value_counts()

brand_counts.plot.bar(color='green', edgecolor='black')

plt.title('Number of Cars by Brand')
plt.xlabel('Brand')
plt.ylabel('Count')

plt.show()


print("Most popular car brand:", brand_counts.idxmax())
```

```python

price_range = pd.cut(car_data['Price'], bins=10)
most_common_price_range = price_range.value_counts().idxmax()
print(f"The most common price range is: {most_common_price_range}")
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Create a scatterplot with a trendline
sns.lmplot(data=car_data, x='Horsepower', y='Mileage', hue='Fuel Type', palette='Set2', ci=None)

# Add labels and title
plt.xlabel('Horsepower')
plt.ylabel('Mileage')
plt.title('Horsepower vs. Mileage')

# Show the plot
plt.show()

# Compute and print the correlation between Horsepower and Mileage
correlation = car_data[['Horsepower', 'Mileage']].corr()
print("Correlation between Horsepower and Mileage:")
print(correlation)
```

```python
import matplotlib.pyplot as plt

fuel_year_counts = car_data.groupby(['Year', 'Fuel Type']).size().unstack()

# Create a grouped bar chart
fuel_year_counts.plot.bar(figsize=(10, 6))

# Add title and labels
plt.title('Fuel Type Preference Over the Years')
plt.xlabel('Manufacturing Year')
plt.ylabel('Number of Cars')
plt.legend(title='Fuel Type')

# Show the plot
plt.show()

# Observations
print("Trend Analysis:")
print(fuel_year_counts)
```
